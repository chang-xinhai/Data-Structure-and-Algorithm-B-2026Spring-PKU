## Problem 3 · 复杂度速判\n\n**题目**：求下面代码的时间复杂度。\n\n**关键点**：先看内层执行次数，再乘外层。

```python
def f(n):\n    s = 0\n    for i in range(n):\n        j = 1\n        while j < n:\n            s += 1\n            j *= 2\n    return s
```

**参考答案**：外层 `n` 次，内层 `log n` 次，总时间复杂度为 $O(n\log n)$。

## Problem 4 · 顺序表插入\n\n**题目**：长度为 `n` 的顺序表，在位置 `i` 插入元素 `x`，写出核心步骤并分析时间复杂度。\n\n**关键点**：先腾位置，再写入。\n\n**解题提要**：从后往前搬移 `i..n-1`，再写入 `x`，时间复杂度 $O(n)$。

```python
def insert(arr, i, x):\n    arr.append(None)\n    for k in range(len(arr) - 1, i, -1):\n        arr[k] = arr[k - 1]\n    arr[i] = x\n    return arr
```

```python
def remove_all(nums, x):\n    w = 0\n    for v in nums:\n        if v != x:\n            nums[w] = v\n            w += 1\n    return nums[:w]
```
