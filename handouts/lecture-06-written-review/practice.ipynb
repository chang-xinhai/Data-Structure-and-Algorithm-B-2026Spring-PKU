## Problem 2 · 复杂度手推

分析下列程序的时间复杂度与空间复杂度：

```python
ans = 0
for i in range(n):
    j = n
    while j > 1:
        ans += 1
        j //= 2
```

要求：写出每层循环的次数来源，而不是只写结论。

### Reference Sketch

- 外层执行 `n` 次。
- 内层每次将 `j` 减半，因此执行 `Theta(log n)` 次。
- 总时间复杂度 `Theta(n log n)`，额外空间 `O(1)`。

## Problem 8 · Dijkstra 手推

对下图从源点 `S` 运行 Dijkstra，写出每轮选入的顶点和 `dist` 更新结果。

- `S->A:1`
- `S->B:4`
- `A->B:2`
- `A->C:5`
- `B->C:1`
- `B->D:6`
- `C->D:3`

````python
graph = {
    'S': {'A': 1, 'B': 4},
    'A': {'B': 2, 'C': 5},
    'B': {'C': 1, 'D': 6},
    'C': {'D': 3},
    'D': {},
}

# 可用这段代码核对你的手算结果
import heapq

def dijkstra(g, s):
    dist = {v: float('inf') for v in g}
    dist[s] = 0
    pq = [(0, s)]
    done = set()
    while pq:
        d, u = heapq.heappop(pq)
        if u in done:
            continue
        done.add(u)
        for v, w in g[u].items():
            if dist[v] > d + w:
                dist[v] = d + w
                heapq.heappush(pq, (dist[v], v))
    return dist

dijkstra(graph, 'S')
````
